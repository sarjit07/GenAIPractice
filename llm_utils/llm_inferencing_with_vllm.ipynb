# vLLM
- vLLM is a high-throughput inference engine (**continuous batching, paged attention, GPU kernel optimization**) for serving LLMs, not just a one-off generate() call.
- Why this beats calling LLM(...) per-request: the ~8s boot cost from the trace we just walked through happens once at server startup, not per request; multiple requests get batched together automatically by the scheduler (that's the 521x concurrency headroom you saw); and you get streaming, retries, and a stable API for free instead of hand-rolling them.
- For a Application, vLLM ships a built-in server that keeps the engine alive and exposes an OpenAI-compatible HTTP API - **vllm serve**

This is how you'd build the application:
```mermaid
flowchart LR
    A[Clients<br/>Web UI, API, cURL]
    B[vLLM API Server <br/> vllm serve]
    C[Scheduler<br/>Continuous Batching]
    D[Inference Engine <br/>KV Cache <br/> vllm]
    E[LLM Model]
    F[GPU Cluster]

    A -->|POST /v1/chat/completions| B
    B --> C
    C --> D
    D --> E
    E --> F

    F -->|Generated Tokens| A
```


````python
import subprocess
import sys
import threading

vllm_bin = str(Path(sys.executable).parent / "vllm")
log_path = Path.cwd() / "vllm_serve.log"
log_file = open(log_path, "w")

server_proc = subprocess.Popen(
    [
        vllm_bin, "serve", MODEL_DIR,
        "--served-model-name", "tinyllama",
        "--port", "8000",
    ],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True,
    bufsize=1,
)

ready = threading.Event()

def _stream_until_ready():
    for line in server_proc.stdout:
        log_file.write(line)
        print(line, end="")
        if "Application startup complete" in line or "Uvicorn running on" in line:
            ready.set()
            break

def _drain_to_file():
    # After startup, keep draining stdout to the log file (not the notebook)
    # so the pipe never fills up and blocks the server.
    for line in server_proc.stdout:
        log_file.write(line)

threading.Thread(target=_stream_until_ready, daemon=True).start()
startup_ok = ready.wait(timeout=180)
threading.Thread(target=_drain_to_file, daemon=True).start()

print(f"\n{'Ready' if startup_ok else 'Timed out waiting'} — pid={server_proc.pid}")
print(f"Further logs -> {log_path}")
print("Stop it later with: server_proc.terminate()")
````
